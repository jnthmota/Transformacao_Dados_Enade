# Extração dos dados do Enade

```python
# Importar bibliotecas necessárias
import pandas as pd #biblioteca de dataframe do pandas
import numpy as np #biblioteca de implementação cientifica , calculos, estatisticas, etc ...
import zipfile #biblioteca para trablhar com arquivos zipados
import requests #biblioteca para trabalhar com links, fazer uma resquest de um determinado link
from io import BytesIO #biblioteca para configurar os inputs dos dados, para trabalhar com ambiente python
import os #biblioteca do sistema, trabalhar com caminhos path
```

```python
# Criando um diretorio para armazernar o conteúdo do ENADE
os.makedirs('./enade2019', exist_ok=True)
```

```python
# Define a url 
url = "https://download.inep.gov.br/microdados/Enade_Microdados/microdados_enade_2019.zip"
# Faz o download do contéudo
filebytes = BytesIO(
    requests.get(url).content    
)

#Extrair o contéudo do zipfile
myzip = zipfile.ZipFile(filebytes)
myzip.extractall("./enade2019")
```

### Verificando os dados baixados

```python
enade = pd.read_csv(
    "./enade2019/3.DADOS/microdados_enade_2019.txt",
    sep = ";", decimal = ",", na_values={"NA", " "}
)
```
